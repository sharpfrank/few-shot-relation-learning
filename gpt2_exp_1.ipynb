```python
s0 = 'The dog is fat.'
s1 = 'His time is short.'
s2 = 'An enormous hound'
s3 = 'A large fat dog'
```

```python
def sent_dists(sents, model, tokenizer):
    toks = tokenizer(sents, padding=True, return_tensors="pt")
    outputs = model(**toks).last_hidden_state
    print(f'{outputs.shape}')
    n_sent = len(sents)
    min_dist = 10000
    min_pair = (-1,-1)
    for i in range(n_sent):
        for j in range(n_sent):
            if i < j:
                print(f'{i}{j}')
                #v1 = torch.mean(outputs[i], dim=1)
                #v2 = torch.mean(outputs[j], dim=1)
                v1 = outputs[i,-1]
                v2 = outputs[j,-1]
                d = torch.dist(v1, v2, p=2)
                print(f'd({i},{j})={d}')
                if d < min_dist:
                    min_dist = d
                    min_pair = (i,j)
    print(f'{min_pair}')
```

```python
sent_dists([s0,s1,s2,s3], model, tokenizer)
```

```python
def pad_sentences(sents):
    sxx = []
    for s in sents:
        sx = s + ' [CLS]'
        sxx.append(sx)
    return sxx
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.add_special_tokens({'pad_token': '[PAD]', 'cls_token': '[CLS]'})
model.resize_token_embeddings(len(tokenizer))
new_s = pad_sentences([s1,s2,s3])
toks = tokenizer(new_s, padding=True, return_tensors="pt")
```
